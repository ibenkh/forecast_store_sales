```python
import pandas as pd 
import numpy as np
import lightgbm as lgb
import datetime
import pickle
```

```python
path = '/Users/idris/Documents/ds_project/forecast_store_sales_kaggle/data/'

cutoff = '2017-08-16'
```

```python
df_cutoff = pd.read_csv(path + 'fe/cutoff/{}/final_frame.csv'.format(cutoff), sep=';').drop('transactions', axis=1)
df_cutoff['date'] = pd.to_datetime(df_cutoff['date'])
```

```python
df_cutoff[df_cutoff['ts_id'] == '9_28'].tail()
```

```python
lgb_params = {
                    'boosting_type': 'gbdt',
                    'objective': 'tweedie',
                    'tweedie_variance_power': 1.1,
                    'metric': 'rmse',
                    'subsample': 0.5,
                    'subsample_freq': 1,
                    'learning_rate': 0.015,
                    'num_leaves': 2**11-1,
                    'min_data_in_leaf': 2**12-1,
                    'feature_fraction': 0.5,
                    'max_bin': 100,
                    'n_estimators': 3000,
                    'boost_from_average': False,
                    'verbose': -1,
                }

horizon = 16
```

```python
cf = '2017-07-31'
```

```python
print(cf)
d_cutoff = pd.to_datetime(cf)
d_cutoff_next = d_cutoff +  datetime.timedelta(days=1)

day_max = d_cutoff +  datetime.timedelta(days=horizon -1)
df_cutoff_t = df_cutoff[df_cutoff['date'] <= day_max]
```

```python
df_cutoff_t.head()
```

```python
df_cutoff_t.shape, df_cutoff_t.id.nunique()
```

```python
X_train = df_cutoff_t[df_cutoff_t['date'] < d_cutoff].drop(['date', 'sales', 'ts_id'], axis=1)
X_val = df_cutoff_t[(df_cutoff_t['date'] >= d_cutoff) & (df_cutoff_t['date'] <= d_cutoff_next)].drop(['date', 'sales','ts_id'], axis=1)

y_train, y_val = df_cutoff_t[df_cutoff_t['date'] < d_cutoff]['sales'], df_cutoff_t[(df_cutoff_t['date'] >= d_cutoff) & (df_cutoff_t['date'] <= d_cutoff_next)]['sales']
```

```python
X_train.shape, y_train.shape
```

```python
X_val.shape, y_val.shape
```

```python
train_data = lgb.Dataset(X_train, label=y_train)
```

```python
lgb_params = {
                    'boosting_type': 'gbdt',
                    'objective': 'tweedie',
                    'tweedie_variance_power': 1.1,
                    'metric': 'rmse',
                    'n_estimators': 500,
                    'verbose': -1,
                }

horizon = 16
```

```python
estimator = lgb.train(lgb_params,
                      train_data,
                      verbose_eval = 100,
                          )
    
display(pd.DataFrame({'name':estimator.feature_name(),
                      'imp':estimator.feature_importance()}).sort_values('imp',ascending=False).head(25))
```

```python
y_pred = estimator.predict(X_val)
```
